```python
import re

def gpt35_token_estimate(text):
    """
    GPT-3.5의 토큰화 방식을 근사하여 토큰 수를 추정하는 함수.
    - 영어 단어, 숫자, 특수문자, 문장 부호 등을 개별 토큰으로 처리.
    - 한글이나 공백도 고려함.
    """
    # 특수문자, 단어, 숫자를 모두 개별 토큰으로 분리
    tokens = re.findall(r"\w+|[^\w\s]|[\s]+", text)
    
    # 공백도 하나의 토큰으로 취급
    tokens = [token for token in tokens if token.strip() != ""]
    
    return len(tokens), tokens

# 예제 데이터
data = """['수성못 유원지',
 '42202',
 nan,
 nan,
 '대구광역시 수성구 용학로 35-5',
 35.826841857699996,
 128.6128734507,
 '수성유원지는 범물동 용지봉(629ｍ)에서 북서부로 뻗어 내린 줄기의 하부에 위치하여 주변의 완만한 산지와 수성못을 끼고 자연의 경관미를 이루고 있다. 1925년에 조성된 수성못을 중심으로 형성되었으며 저수량도 70만에 달하고 21만 8000㎡의 면적에 못 둘레도 2,020m에 이른다. 수성못은 1925년 일제강점기에 농업용수 공급용으로 조성된 인공 못이었다. 지금은 수변 휴식공간으로 활용되어 연중 풍부한 수량을 자랑하고 있다. 수성못 주변에는 벤치와 수목, 산책로 유람선 선착장, 우주선, 유람선, 바이킹, 범버카, 회전목마 등의 놀이시설을 갖춘 수성랜드가  어우러져 유원지를 형성하고 있다. 보트놀이와 오리배 등을 즐길 수 있는 수성못과 어린이 놀이터, 두산폭포 등 볼거리, 즐길거리가 있어 가족나들이와 연인들의 데이트 코스로 적격이다. 수성못 미디어아트 음악분수는 이색적인 볼거리를 제공한다.',
 nan,
 '수성유원지 053-666-2863<br/>\n수성못 관광안내소 053-761-0645',
 nan,
 '연중무휴 (00:00~24:00)',
 nan,
 '전연령',
 nan,
 nan,
 '00:00~24:00<br>\r\n※ 수성 미디어아트 음악분수 운영시간은 홈페이지 참조 요망',
 '가능',
 '없음',
 nan,
 nan,
 '이용가능시설:- 수성못, 오리배&유람선, 수변데크로드, 포켓무대, 관광안내소 북카페, 상화동산, 시문학거리, 들안길 시화거리, 영상음악분수, 아르떼 수성랜드, 수성파크랜드<br />\r\n<br />\r\n※ 기상여건 및 사정에따라 이용불가시설 있음\n화장실:있음\n']
추정 """

# 함수 실행
token_count, tokens = gpt35_token_estimate(data)

print(f"추정 토큰 수: {token_count}")
print(f"토큰 예시: {tokens[:20]}")  # 첫 20개의 토큰만 출력
```

```python
import openai
import json
from tqdm import tqdm
from dotenv import load_dotenv

# 환경 변수 로드 (API 키 및 기타 설정)
load_dotenv(dotenv_path="/home/hwang/projects/yanolja/src/.env")

# JSONL 파일 경로 설정
input_file_path = "/home/hwang/projects/yanolja/src/modeling/data/instruction_dataset_all_rows_all_scenarios_gpt35_tourAPI_tourist_attractions_seoul_gyeonggi_2.4K_scenario.jsonl"
output_file_path = "instruction_dataset_all_rows_all_scenarios_gpt35_tourAPI_tourist_attractions_seoul_gyeonggi_2.4K_scenario_augmented.jsonl"

def augment_text(original_text, text_type="instruction", num_augments=5):
    """
    instruction, input, output 각각의 의미를 유지하면서 다양한 형태로 변형하는 함수.
    """
    augments = []
    # 각 text_type에 맞는 프롬프트 설정
    prompts = {
        "instruction": {
            "system_message": "다음 instruction 문장을 의미를 유지하면서 다양한 표현으로 변경하세요. 완성된 문장만 작성하세요.",
            "few_shot_example": (
                "예제:\n"
                "원문: 사용자가 관광지의 기본 정보를 요청했을 때 답변을 생성하세요.\n"
                "관광지 정보를 요청받은 경우 적절히 답변을 작성하세요."
            ),
        },
        "input": {
            "system_message": "다음 질문 텍스트를 자연스럽고 다양한 표현 방식으로 변경하세요. 질문만 작성하세요.",
            "few_shot_example": (
                "예제:\n"
                "원문: 연천 초성김치마을의 기본 정보를 알려주세요.\n"
                "연천 초성김치마을은 어떤 곳인가요?"
            ),
        },
        "output": {
            "system_message": "다음 응답 텍스트를 의미를 유지하면서 다양한 표현으로 변경하세요. 완전한 문장만 작성하세요.",
            "few_shot_example": (
                "예제:\n"
                "원문: 연천 초성김치마을은 경기 연천에 위치한 전통 김치 체험 공간입니다.\n"
                "연천 초성김치마을은 경기도 연천에 있는 전통 김치 체험 장소입니다."
            ),
        },
    }

    # 선택된 text_type에 맞는 프롬프트를 가져옴
    if text_type not in prompts:
        raise ValueError("text_type은 'instruction', 'input', 'output' 중 하나여야 합니다.")

    system_message = prompts[text_type]["system_message"]
    few_shot_example = prompts[text_type]["few_shot_example"]

    try:
        # OpenAI API 호출 (한 번에 5개 생성)
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo-0125",
            messages=[
                {"role": "system", "content": system_message},
                {"role": "user", "content": f"{few_shot_example}\n\n원문: {original_text}\n결과:"},
            ],
            temperature=0.7,
            n=num_augments,  # 한 번에 생성할 변형 수
        )

        # 응답에서 모든 변형 텍스트 추출
        for choice in response["choices"]:
            new_text = choice["message"]["content"].strip()
            # 숫자나 불필요한 포맷 제거
            new_text = new_text.lstrip("1. ").lstrip("2. ").strip()
            augments.append(new_text)

    except Exception as e:
        # 에러 발생 시 출력
        print(f"Error during augmentation: {e}")

    return augments


# JSONL 파일 읽기 함수
def read_jsonl(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        for line in file:
            yield json.loads(line)

# JSONL 파일 쓰기 함수
def write_jsonl(file_path, data):
    with open(file_path, "a", encoding="utf-8") as file:
        for item in data:
            file.write(json.dumps(item, ensure_ascii=False) + "\n")

# 데이터셋 증강 함수
def augment_dataset(input_path, output_path):
    data = read_jsonl(input_path)
    for item in tqdm(data, desc="Processing"):
        original_instruction = item["instruction"]
        original_input = item["input"]

        # instruction 및 input 필드 증강
        augmented_instructions = augment_text(original_instruction, text_type="instruction")
        augmented_inputs = augment_text(original_input, text_type="input")
        
        # 증강된 데이터를 새로운 형태로 저장
        augmented_data = [
            {
                "instruction": aug_inst,
                "input": aug_input,
                "output": item["output"],
            }
            for aug_inst, aug_input in zip(augmented_instructions, augmented_inputs)
        ]
        # 실시간 저장
        write_jsonl(output_path, augmented_data)

# 데이터 증강 실행
augment_dataset(input_file_path, output_file_path)
```
