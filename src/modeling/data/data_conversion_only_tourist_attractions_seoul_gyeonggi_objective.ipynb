```python
import re

def gpt35_token_estimate(text):
    """
    GPT-3.5의 토큰화 방식을 근사하여 토큰 수를 추정하는 함수.
    - 영어 단어, 숫자, 특수문자, 문장 부호 등을 개별 토큰으로 처리.
    - 한글이나 공백도 고려함.
    """
    # 특수문자, 단어, 숫자를 모두 개별 토큰으로 분리
    tokens = re.findall(r"\w+|[^\w\s]|[\s]+", text)
    
    # 공백도 하나의 토큰으로 취급
    tokens = [token for token in tokens if token.strip() != ""]
    
    return len(tokens), tokens

# 예제 데이터
data = """['수성못 유원지',
 '42202',
 nan,
 nan,
 '대구광역시 수성구 용학로 35-5',
 35.826841857699996,
 128.6128734507,
 '수성유원지는 범물동 용지봉(629ｍ)에서 북서부로 뻗어 내린 줄기의 하부에 위치하여 주변의 완만한 산지와 수성못을 끼고 자연의 경관미를 이루고 있다. 1925년에 조성된 수성못을 중심으로 형성되었으며 저수량도 70만에 달하고 21만 8000㎡의 면적에 못 둘레도 2,020m에 이른다. 수성못은 1925년 일제강점기에 농업용수 공급용으로 조성된 인공 못이었다. 지금은 수변 휴식공간으로 활용되어 연중 풍부한 수량을 자랑하고 있다. 수성못 주변에는 벤치와 수목, 산책로 유람선 선착장, 우주선, 유람선, 바이킹, 범버카, 회전목마 등의 놀이시설을 갖춘 수성랜드가  어우러져 유원지를 형성하고 있다. 보트놀이와 오리배 등을 즐길 수 있는 수성못과 어린이 놀이터, 두산폭포 등 볼거리, 즐길거리가 있어 가족나들이와 연인들의 데이트 코스로 적격이다. 수성못 미디어아트 음악분수는 이색적인 볼거리를 제공한다.',
 nan,
 '수성유원지 053-666-2863<br/>\n수성못 관광안내소 053-761-0645',
 nan,
 '연중무휴 (00:00~24:00)',
 nan,
 '전연령',
 nan,
 nan,
 '00:00~24:00<br>\r\n※ 수성 미디어아트 음악분수 운영시간은 홈페이지 참조 요망',
 '가능',
 '없음',
 nan,
 nan,
 '이용가능시설:- 수성못, 오리배&유람선, 수변데크로드, 포켓무대, 관광안내소 북카페, 상화동산, 시문학거리, 들안길 시화거리, 영상음악분수, 아르떼 수성랜드, 수성파크랜드<br />\r\n<br />\r\n※ 기상여건 및 사정에따라 이용불가시설 있음\n화장실:있음\n']
추정 """

# 함수 실행
token_count, tokens = gpt35_token_estimate(data)

print(f"추정 토큰 수: {token_count}")
print(f"토큰 예시: {tokens[:20]}")  # 첫 20개의 토큰만 출력
```

```python
## raw dataset to qa dataset using gpt3.5

import openai
import pandas as pd
import json
from tqdm import tqdm

import os
from dotenv import load_dotenv

load_dotenv()

# OpenAI API 키 설정
openai.api_key = os.getenv("OPENAI_API_KEY")

data_path = "yanolja/data/dataset/tourAPI/touristAttractions.xls" # tourist_attraction 으로 제한
df = pd.read_excel(data_path)
df = df[df['주소'].str.contains('서울특별시|경기도', na=False)] # 서울특별시, 경기도로 제한

error_log_path = "qa_generation_errors.log"

qa_dataset = []

def generate_qa_prompt(row):
    prompt = f"""
    다음은 관광지 데이터입니다. 이 데이터를 바탕으로 **객관적 사실 기반 QA**와 **추천 조건 기반 QA**를 가능한 많이 생성해주세요.
    ---

    ### **지침**

    1. **Raw Data에 있는 내용만 활용하세요.**  
      - **Raw Data**에 포함되지 않은 정보는 "제공된 데이터에 정보가 없습니다"라고 명확하게 답변하세요.  
      - 예시:  
        - 질문: "명성산에서 유모차를 대여할 수 있나요?"  
        - 답변: "제공된 데이터에 명성산의 유모차 대여 정보는 없습니다."

    2. **모든 질문에는 관광지 명칭이 포함되어야 합니다.**  
      - 질문이 자연스럽고 구체적이되, 반드시 **관광지 명칭**을 명시하세요.  
        예시:  
        - "명성산에는 주차장이 있나요?"  
        - "명성산의 등산로는 어떤 코스가 있어요?"

    3. **같은 필드라도 다양한 시각에서 질문을 작성하세요.**
      - 예를 들어, 체험 프로그램 필드를 사용해 다양한 질문을 만드세요:  
        - "연천 초성김치마을에 가면 어떤 체험 프로그램을 추천하시나요?"  
        - "연천 초성김치마을에서 김치 만들기 체험 비용은 얼마인가요?"  
        - "연천 초성김치마을의 체험 프로그램은 예약이 필요한가요?"

    4. **답변은 명확하고 구체적으로 작성하세요.**  
      - Raw Data의 정보를 바탕으로 친절하고 유용하게 작성하되, 불확실하거나 없는 정보는 명확하게 언급하세요.  

    5. **각 필드를 기반으로 다양한 질문과 답변을 생성하세요.**  
      - **개요**: 관광지의 특색, 주요 활동에 대해 질문하세요.  
        - 질문 예시: "명성산은 어떤 곳인가요?"  
      - **상세정보**: 등산로, 체험 안내, 주변 경관에 대한 질문을 만드세요.  
        - 질문 예시: "명성산에는 어떤 등산로가 있나요?"  
      - **이용시간/이용시기**: 방문하기 좋은 시간이나 추천 계절에 대해 질문하세요.  
        - 질문 예시: "명성산을 방문하기 좋은 계절은 언제인가요?"  
      - **시설**: 주차장, 화장실, 편의시설 등 방문자에게 도움이 될 질문을 작성하세요.  
        - 질문 예시: "명성산에는 주차 시설이 있나요?"  

    6. **필드 내용을 기반으로 질문을 다양화하고 구체화하세요.**  
      - 질문을 만들 때 특정 상황이나 대상을 추가해 질문을 세분화하세요.  
        예시:  
        - "명성산은 가족과 함께 가기에 어떤 점이 좋은가요?"  
        - "명성산을 가을에 방문하면 어떤 경험을 할 수 있나요?"  
        - "명성산 등산로 중 초보자가 가기 쉬운 코스는 어디인가요?"

    7. **프로그램별 맞춤 추천 및 세분화된 답변을 작성하세요.**
      - **프로그램별 비용, 예약 방법, 대상 적합성**을 반영하세요.
      - 질문 예시:
        - "연천 초성김치마을에서 김치 만들기 체험은 어떻게 진행되나요?"
        - "연천 초성김치마을의 체험 프로그램 비용은 얼마인가요?"
        - "연천 초성김치마을은 가족과 함께 체험하기 좋은 프로그램이 있나요?"

    8. **추천 조건에 맞춰 다양한 상황을 반영하세요.**
       - **대상별**: 가족, 연인, 혼자 방문, 어르신
       - **시간대별**: 오전, 오후, 저녁
       - **계절별**: 봄, 여름, 가을, 겨울
       - **목적별**: 체험, 휴식, 사진 명소, 편의시설

    9. **추천 조건에 따른 QA 작성 시, 데이터와 일치하도록 작성하세요.**  
      - **추천 조건**: 가족, 연인, 계절, 시간대 등과 같은 범주를 활용하세요.  
      - 예시:  
        - "명성산은 가을에 방문하기 좋은 장소인가요?"  
          - 답변: "명성산은 가을철 억새꽃 산행으로 유명하며, 아름다운 단풍을 감상할 수 있습니다."  

    10. **자연스럽고 일상적인 질문을 작성하세요.**  
      - 일반인이 질문하는 방식처럼 대화체를 반영하세요.  
        예시:  
        - "가족이랑 주말에 명성산 가려고 하는데, 어떤 체험을 할 수 있어요?"  
        - "명성산은 혼자 가기 좋은 곳인가요?"

    ---

    ### **관광지 데이터**
    {f"- 명칭: {row['명칭']}" if pd.notna(row['명칭']) else ''}
    {f"- 주소: {row['주소']}" if pd.notna(row['주소']) else ''}
    {f"- 개요: {row['개요']}" if pd.notna(row['개요']) else ''}
    {f"- 유산구분: {row[' 유산구분']}" if pd.notna(row[' 유산구분']) else ''}
    {f"- 쉬는날: {row['쉬는날']}" if pd.notna(row['쉬는날']) else ''}
    {f"- 이용시기: {row['이용시기']}" if pd.notna(row['이용시기']) else ''}
    {f"- 이용시간: {row['이용시간']}" if pd.notna(row['이용시간']) else ''}
    {f"- 주차시설: {row['주차시설']}" if pd.notna(row['주차시설']) else ''}
    {f"- 체험안내: {row['체험안내']}" if pd.notna(row['체험안내']) else ''}    
    {f"- 체험가능연령: {row['체험가능연령']}" if pd.notna(row['체험가능연령']) else ''}
    {f"- 유모차 대여 여부: {row['유모차 대여 여부']}" if pd.notna(row['유모차 대여 여부']) else ''}
    {f"- 애완동물 동반 가능 여부: {row['애완동물 동반 가능 여부']}" if pd.notna(row['애완동물 동반 가능 여부']) else ''}
    {f"- 상세정보: {row['상세정보']}" if pd.notna(row['상세정보']) else ''}


    ---

    ### **객관적 사실 기반 QA 예시**
    - Q: "연천 초성김치마을에서 김치 만들기 체험 비용은 얼마인가요?"
      A: "김치 만들기 체험은 1인당 15,000원입니다."

    - Q: "연천 초성김치마을의 숙박 시설은 어떤 종류가 있나요?"
      A: "연천 초성김치마을에는 25평형 단독형 펜션 5동, 13평형 원룸형 펜션 4동, 40평형 빌라형 펜션 1동이 마련되어 있습니다."

    ### **추천 조건 기반 QA 예시**
    - Q: "봄에 방문하기 좋은 체험 마을이 있나요?"
      A: "연천 초성김치마을은 봄철에 꽃차 체험과 산책로를 따라 봄 경치를 즐길 수 있어 봄에 방문하기 좋은 장소입니다."

    - Q: "연천 초성김치마을은 가족과 함께 방문하기 좋은가요?"
      A: "연천 초성김치마을은 어린이놀이터와 다양한 체험 프로그램을 제공하여 가족 단위 방문객에게 적합한 장소입니다."

    ---

    **출력 형식**:
    {{
      "objective_qa": [
        {{
          "question": "객관적 사실에 기반한 질문",
          "answer": "답변"
        }},
        ...
      ],
      "recommendation_qa": [
        {{
          "question": "추천 조건에 맞는 질문",
          "answer": "추천 답변"
        }},
        ...
      ]
    }}
    """
    return prompt

def generate_qa(row):
    """GPT-3.5를 호출하여 QA를 생성하는 함수"""
    try:
        prompt = generate_qa_prompt(row)
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo-0125",
            messages=[
                {"role": "system", "content": "당신은 관광 정보에 대한 자연스럽고 일상적이며 유용한 QA 데이터를 생성하는 도우미입니다. 질문은 상세하고 자연스럽게 작성하며, 답변은 제공된 데이터를 기반으로 유용하고 구체적인 정보를 포함해주세요. 질문과 답변은 조금 길어도 좋지만 주어진 데이터를 벗어나지 마세요. 예를 들어 '명성산 등산로는 어떤 경치를 볼 수 있나요?' 같은 자연스러운 질문과 답변을 생성하세요."},
                {"role": "user", "content": prompt}],
            max_tokens=4096,
            temperature=0
        )
        return json.loads(response['choices'][0]['message']['content'])
    except Exception as e:
        return {"error": str(e)}
    
# 결과 저장
batch_size = 50  # 중간 저장 배치 크기
error_count = 0

output_path = "qa_dataset_only_tourist_attractions_seoul_gyeonggi.json"
with open(output_path, "a", encoding="utf-8") as output_file, open(error_log_path, "a") as error_log:
    for i, (_, row) in enumerate(tqdm(df.iterrows(), total=len(df))):
        try:
            result = generate_qa(row)
            if "error" in result:
                error_log.write(f"Error at row {i} ({row['명칭']}): {result['error']}\n")
                error_count += 1
                continue

            result_with_name = {
                "name": row['명칭'] if pd.notna(row['명칭']) else "알 수 없는 명칭",
                "qa": result
            }
            output_file.write(json.dumps(result_with_name, ensure_ascii=False) + "\n")

            # 중간 저장
            if (i + 1) % batch_size == 0:
                output_file.flush()

        except Exception as e:
            error_log.write(f"Unexpected Error at row {i} ({row['명칭']}): {str(e)}\n")
            error_count += 1
            continue

print(f"QA 데이터셋이 '{output_path}'에 저장되었습니다. 에러 수: {error_count}")
```
