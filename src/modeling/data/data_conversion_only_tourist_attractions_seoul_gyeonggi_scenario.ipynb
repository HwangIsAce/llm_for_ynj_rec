```python
import re

def gpt35_token_estimate(text):
    """
    GPT-3.5의 토큰화 방식을 근사하여 토큰 수를 추정하는 함수.
    - 영어 단어, 숫자, 특수문자, 문장 부호 등을 개별 토큰으로 처리.
    - 한글이나 공백도 고려함.
    """
    # 특수문자, 단어, 숫자를 모두 개별 토큰으로 분리
    tokens = re.findall(r"\w+|[^\w\s]|[\s]+", text)
    
    # 공백도 하나의 토큰으로 취급
    tokens = [token for token in tokens if token.strip() != ""]
    
    return len(tokens), tokens

# 예제 데이터
data = """['수성못 유원지',
 '42202',
 nan,
 nan,
 '대구광역시 수성구 용학로 35-5',
 35.826841857699996,
 128.6128734507,
 '수성유원지는 범물동 용지봉(629ｍ)에서 북서부로 뻗어 내린 줄기의 하부에 위치하여 주변의 완만한 산지와 수성못을 끼고 자연의 경관미를 이루고 있다. 1925년에 조성된 수성못을 중심으로 형성되었으며 저수량도 70만에 달하고 21만 8000㎡의 면적에 못 둘레도 2,020m에 이른다. 수성못은 1925년 일제강점기에 농업용수 공급용으로 조성된 인공 못이었다. 지금은 수변 휴식공간으로 활용되어 연중 풍부한 수량을 자랑하고 있다. 수성못 주변에는 벤치와 수목, 산책로 유람선 선착장, 우주선, 유람선, 바이킹, 범버카, 회전목마 등의 놀이시설을 갖춘 수성랜드가  어우러져 유원지를 형성하고 있다. 보트놀이와 오리배 등을 즐길 수 있는 수성못과 어린이 놀이터, 두산폭포 등 볼거리, 즐길거리가 있어 가족나들이와 연인들의 데이트 코스로 적격이다. 수성못 미디어아트 음악분수는 이색적인 볼거리를 제공한다.',
 nan,
 '수성유원지 053-666-2863<br/>\n수성못 관광안내소 053-761-0645',
 nan,
 '연중무휴 (00:00~24:00)',
 nan,
 '전연령',
 nan,
 nan,
 '00:00~24:00<br>\r\n※ 수성 미디어아트 음악분수 운영시간은 홈페이지 참조 요망',
 '가능',
 '없음',
 nan,
 nan,
 '이용가능시설:- 수성못, 오리배&유람선, 수변데크로드, 포켓무대, 관광안내소 북카페, 상화동산, 시문학거리, 들안길 시화거리, 영상음악분수, 아르떼 수성랜드, 수성파크랜드<br />\r\n<br />\r\n※ 기상여건 및 사정에따라 이용불가시설 있음\n화장실:있음\n']
추정 """

# 함수 실행
token_count, tokens = gpt35_token_estimate(data)

print(f"추정 토큰 수: {token_count}")
print(f"토큰 예시: {tokens[:20]}")  # 첫 20개의 토큰만 출력
```

```python
import json
from datasets import Dataset

import os
from dotenv import load_dotenv

# 데이터 평탄화 함수
def flatten_dict(d, parent_key='', sep='_'):
    items = []
    for k, v in d.items():
        new_key = f"{parent_key}{sep}{k}" if parent_key else k
        if isinstance(v, dict):
            items.extend(flatten_dict(v, new_key, sep=sep).items())
        else:
            items.append((new_key, v))
    return dict(items)

file_path = "src/modeling/data/instruction_dataset_all_rows_all_scenarios_gpt35_tourAPI_tourist_attractions_seoul_gyeonggi_2.4K_scenario.jsonl"

instruction = []
with open(file_path, 'r', encoding='utf-8') as f:
    for line in f:
        try:
            tmp = json.loads(line)
            instruction.append(tmp)
        except json.JSONDecodeError as e:
            print(f"Error decoding line: {e}")

flattened_instruction = [flatten_dict(row) for row in instruction]

dataset = Dataset.from_list(flattened_instruction)

load_dotenv()

hf_token = os.getenv("HF_API_KEY")

dataset_name = "passionMan/ya_tourist_attractions_seoul_gyeonggi_by_scenario_1219"  
dataset.push_to_hub(dataset_name, token=hf_token)
```

```python
import json

# 원본 QA 데이터 로드
qa_data = qa

instruction_templates = [
    "주어진 질문에 대한 정확한 답변을 제공하세요. 답변은 제공된 데이터를 기반으로 작성하고, 추가 정보가 없으면 명확히 알리세요.",
    "제공된 데이터에서 필요한 정보를 추출하고 자연스러운 언어로 답변하세요.",
    "사용자 질문에 따라 조건을 이해하고 답변하세요. 예를 들어, 대상, 시간대, 계절 등을 고려해 답변을 작성하세요.",
    "사용자 질문에 친절하고 자세한 정보를 포함해 답변하세요.",
    "주어진 데이터에 있는 근거를 활용하여 답변을 논리적으로 작성하세요.",
    "사용자의 질문에 대해 구체적이고 정확한 정보를 바탕으로 답변을 작성하세요.",
    "다음 질문에 정확히 답변해 주세요.",
    "사용자의 질문에 친절하고 명확하게 답변하세요."
]

# Output JSONL 파일 경로
# output_file = "instruction_dataset_only_tourist_attractions_seoul_gyeonggi.jsonl"

# JSONL로 저장하는 함수
def save_to_jsonl(data, file_path):
    with open(file_path, "a", encoding="utf-8") as f:
        f.write(json.dumps(data, ensure_ascii=False) + "\n")

# Instruction 생성 및 데이터 변환
with open(output_file, "w", encoding="utf-8") as f:  # 기존 파일 초기화
    pass

for entry in qa_data:
    name = entry['name']
    qa_sections = entry['qa']

    for qa_type, qa_list in qa_sections.items():
        for qa in qa_list:
            question = qa['question']
            answer = qa['answer']

            # 각 question에 대해 여러 instruction 생성
            for template in instruction_templates:
                instruction = template
                result = {
                    "name": name,
                    "instruction": instruction,
                    "input": question,  # question을 input으로 설정
                    "output": answer  # answer를 output으로 설정
                }
                save_to_jsonl(result, output_file)

print(f"수정된 QA 데이터셋이 {output_file}에 저장되었습니다.")
```

```python
from datasets import Dataset

import os
from dotenv import load_dotenv

file_path = "src/modeling/data/instruction_dataset_only_tourist_attractions_seoul_gyeonggi.jsonl"

instruction = []
with open(file_path, 'r', encoding='utf-8') as f:
    for line in f:
        try:
            tmp = json.loads(line)
            instruction.append(tmp)
        except json.JSONDecodeError as e:
            print(f"Error decoding line: {e}")

dataset = Dataset.from_list(instruction)  

load_dotenv()

hf_token = os.getenv("HF_API_KEY")

dataset_name = "passionMan/ya_tourist_attractions_seoul_gyeonggi"  
dataset.push_to_hub(dataset_name, token=hf_token)
```
