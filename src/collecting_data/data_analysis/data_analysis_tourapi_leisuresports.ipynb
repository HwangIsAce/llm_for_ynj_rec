````python
import openai
import pandas as pd
import re
from tqdm import tqdm 

import os
from dotenv import load_dotenv

load_dotenv()

openai.api_key = os.getenv("OPENAI_API_KEY")

def preprocess_day_off(day_off):
    if not day_off or pd.isna(day_off):
        return [], []

    days = []
    special_notes = []

    if "월요일" in day_off:
        days.append("월")
    if "화요일" in day_off:
        days.append("화")
    if "수요일" in day_off:
        days.append("수")
    if "목요일" in day_off:
        days.append("목")
    if "금요일" in day_off:
        days.append("금")
    if "토요일" in day_off:
        days.append("토")
    if "일요일" in day_off:
        days.append("일")
    if "주말" in day_off:
        days.extend(["토", "일"])
    if "공휴일" in day_off:
        days.append("공휴일")
    if "마지막 주 일요일" in day_off:
        days.append("일")
        special_notes.append("마지막 주 일요일 휴무")

    if "전화문의" in day_off:
        special_notes.append("휴무일은 전화문의 필요")

    return list(set(days)), list(set(special_notes))  

def preprocess_available_time(available_time):

    if not available_time or pd.isna(available_time):
        return "", []  

    time_only = re.sub(r"\s*\(.*?\)", "", available_time).strip()
    special_notes = re.findall(r"\(.*?\)", available_time)

    special_notes_cleaned = [note.strip("()").strip() for note in special_notes]
    return time_only, special_notes_cleaned

def generate_template_with_gpt(preprocessed_day_off, time_and_notes):
    available_time, special_notes = time_and_notes
    day_off_days, day_off_notes = preprocessed_day_off

    all_special_notes = special_notes + day_off_notes
    special_notes_str = ", ".join(list(set(all_special_notes))) if all_special_notes else "정보 없음"

    prompt = f"""
    다음은 문화시설의 운영 정보를 나타내는 데이터입니다:
    - 휴무일: {", ".join(day_off_days) or "정보 없음"}
    - 영업시간: {available_time or "정보 없음"}
    - 특이사항: {special_notes_str}
    
    위 정보를 기반으로 아래의 템플릿을 채워주세요:
    - 월: 영업시간
    - 화: 영업시간
    - 수: 영업시간
    - 목: 영업시간
    - 금: 영업시간
    - 토: 영업시간
    - 일: 영업시간
    - 휴무: []
    - 특이사항: []

    주의사항:
    1. 휴무일로 지정된 요일은 "휴무"로 표시하고, 영업시간을 작성하지 마세요.
    2. 영업시간에 특이한 규칙이 있으면 "특이사항"에 추가해주세요.
    3. JSON 형식으로 반환해주세요.
    """

    response = openai.ChatCompletion.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7
    )
    
    reply = response['choices'][0]['message']['content']
    return reply

def process_and_add_column(df):
    results = [] 

    for _, row in tqdm(df.iterrows(), total=len(df), desc="Processing rows", unit="row"):
        day_off = preprocess_day_off(row.get("day_off"))
        available_time = row.get("available_time")
        time_and_notes = preprocess_available_time(available_time)

        try:
            template = generate_template_with_gpt(day_off, time_and_notes)
            results.append(template)
        except Exception as e:
            results.append({"error": str(e)})

    df["available_time_v1"] = results
    return df

tourapi = process_and_add_column(tourapi)

output_csv_path = "tourapi_leisuresports_processed.csv"
tourapi.to_csv(output_csv_path, index=False, encoding="utf-8-sig")

print(f"데이터가 다음 경로에 저장되었습니다: {output_csv_path}")
````

````python
import json
import pandas as pd

def parse_json_from_string(value):
    try:
        if isinstance(value, str) and value.startswith("```json") and value.endswith("```"):
            cleaned_value = value.replace("```json\n", "").replace("```", "").strip()
            return json.loads(cleaned_value)
        else:
            raise ValueError("String is not in the expected format")
    except json.JSONDecodeError as e:
        print(f"JSON decode error: {e}")
        return None  

def add_opening_period_to_details(row):
    try:
        available_time_data = parse_json_from_string(row["available_time_v1"])
        if available_time_data is None:
            raise ValueError("Invalid JSON format in available_time_v1")

        opening_period_info = row["opening_period"]
        if opening_period_info and not pd.isna(opening_period_info):
            if "특이사항" in available_time_data:
                if isinstance(available_time_data["특이사항"], list):
                    available_time_data["특이사항"].append(f"개장기간: {opening_period_info}")
                else:
                    available_time_data["특이사항"] = [
                        available_time_data["특이사항"],
                        f"개장기간: {opening_period_info}"
                    ]
            else:
                available_time_data["특이사항"] = [f"개장기간: {opening_period_info}"]
        else:
            if "특이사항" in available_time_data:
                if isinstance(available_time_data["특이사항"], list):
                    available_time_data["특이사항"].append("개장기간: NaN")
                else:
                    available_time_data["특이사항"] = [
                        available_time_data["특이사항"],
                        "개장기간: NaN"
                    ]
            else:
                available_time_data["특이사항"] = ["개장기간: NaN"]

        return "```json\n" + json.dumps(available_time_data, ensure_ascii=False, indent=2) + "\n```"
    except Exception as e:
        print(f"Error processing row: {e}")
        return row["available_time_v1"]  

tourapi["available_time_v1"] = tourapi.apply(add_opening_period_to_details, axis=1)
````

````python
import pandas as pd
import numpy as np

def combine_columns(row):
    additional_info = []

    # fee 추가
    if pd.notna(row['fee']):
        additional_info.append(f"요금: {row['fee']}")
    
    # stroller 추가
    if pd.notna(row['stroller']):
        additional_info.append(f"유모차: {row['stroller']}")
    
    # pet 추가
    if pd.notna(row['pet']):
        additional_info.append(f"반려동물: {row['pet']}")

    # age_available 추가
    if pd.notna(row['age_available']):
        additional_info.append(f"체험가능연령: {row['age_available']}")
    
    details = row['details'] if pd.notna(row['details']) else ""
    
    combined_details = details + ('\n' if details and additional_info else '') + '\n'.join(additional_info)
    return combined_details

tourapi['details'] = tourapi.apply(combine_columns, axis=1)

tourapi = tourapi.drop(columns=['fee', 'stroller', 'pet', 'age_available'])
````
