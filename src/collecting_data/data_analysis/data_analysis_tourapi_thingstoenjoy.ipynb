```python
import json

# row_index 열 추가
tourapi["row_index"] = tourapi.index

# JSON 생성
output = []

for _, row in tourapi.iterrows():
    # special_conditions 필드 구성
    transformed_data = {
        "special_conditions": [
            {"type": "event_start_date", "details": row["event_start_date"]},
            {"type": "event_end_date", "details": row["event_end_date"]},
            {"type": "performance_time", "details": row["performance_time"]}
        ]
    }
    
    # JSON 객체 추가
    output.append({
        "row_index": row["row_index"],
        "transformed_data": transformed_data
    })

tourapi['available_time_v2'] = output
```

```python
import pandas as pd
import numpy as np

def combine_columns(row):
    additional_info = []

    # organizer_info 추가
    if pd.notna(row['organizer_info']):
        additional_info.append(f"주최자 정보: {row['organizer_info']}")
    
    # organizer_phone 추가
    if pd.notna(row['organizer_phone']):
        additional_info.append(f"주최자 연락처: {row['organizer_phone']}")
    
    # event_location 추가
    if pd.notna(row['event_location']):
        additional_info.append(f"행사장소: {row['event_location']}")
    
    # fee 추가
    if pd.notna(row['fee']):
        additional_info.append(f"이용요금: {row['fee']}")

    # progression_type 추가
    if pd.notna(row['progression_type']):
        additional_info.append(f"진행방식: {row['progression_type']}")

    # festival_type 추가
    if pd.notna(row['festival_type']):
        additional_info.append(f"행사종류: {row['festival_type']}")
    
    details = row['details'] if pd.notna(row['details']) else ""
    
    combined_details = details + ('\n' if details and additional_info else '') + '\n'.join(additional_info)
    return combined_details

tourapi['details'] = tourapi.apply(combine_columns, axis=1)

tourapi = tourapi.drop(columns=['organizer_info', 'organizer_phone', 'event_location', 'fee', 'progression_type',
                                'festival_type'])
```
